```python
import nltk
from nltk.corpus import gutenberg
from collections import Counter
import matplotlib.pyplot as plt
import numpy as np

# Assurez-vous d'avoir téléchargé le corpus Gutenberg
nltk.download('gutenberg')

# 1. Charger le texte de "Ulysse" de James Joyce
# Nous utilisons un texte du corpus Gutenberg pour l'exemple
texte = gutenberg.words('austen-emma.txt')  # Remplacer par 'ulysses.txt' si disponible
mots = [mot.lower() for mot in texte if mot.isalpha()]

# 2. Compter les occurrences de chaque mot
frequence_mots = Counter(mots)
mots_ordonne = frequence_mots.most_common()  # Liste de (mot, fréquence) par ordre décroissant

# 3. Calculer f(n) * n pour chaque rang et vérifier la constance de la valeur
rangs = np.arange(1, len(mots_ordonne) + 1)  # Rangs des mots
frequences = np.array([freq for _, freq in mots_ordonne])  # Fréquences des mots
valeurs_zipf = frequences * rangs  # Calcul de f(n) * n

# 4. Afficher la distribution de Zipf
plt.figure(figsize=(10, 6))
plt.plot(rangs, valeurs_zipf, label="f(n) * n")
plt.axhline(y=np.mean(valeurs_zipf), color='r', linestyle='--', label="Constante estimée K")
plt.xlabel("Rang n")
plt.ylabel("f(n) * n")
plt.title("Vérification de la loi de Zipf")
plt.legend()
plt.show()
```
